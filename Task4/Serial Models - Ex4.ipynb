<u> A STANDARD QUESTION ON MARKOV CHAINS </u>

#### Consider a Markov chain with states $S=\{0;\,\dots;\,N\}$ and transition probabilities $p_{i;\,i+1}=p;\,p_{i;\,i-1}=q$ for $1\leq i\leq N-1$, where $p+q=1$; assume $p_{0;\,1}=1; \, p_{N;\,N-1}=1$.

#### 1. Draw the transition diagram.

The example diagram provided for N=5 serves as a representative illustration, demonstrating the system's behavior for any N.

#### 2. Is the Markov chain irreducible?


Yes. Upon analyzing the transition diagram, we can see that every state can be reached from every other state, meaning that the markov chain is irreducible.

#### 3. Find the stationary distribution.

Let's define the transition matrix of the Markov Chain as follows:

$$
P = \begin{bmatrix}
0 & 1 & 0 & \cdots & 0 & 0 \\
1-p & 0 & p & \cdots & 0 & 0 \\
0 & 1-p & 0 & \cdots & 0 & \vdots \\
\vdots & \vdots & \vdots & \ddots & \vdots & 0 \\
0 & 0 & 0 & \cdots & 1-p & p \\
0 & 0 & 0 & \cdots & 1 & 0 \\
\end{bmatrix}
$$

Based on the transition matrix and the diagram, we can establish the relationships between the stationary probabilities of the states:

$$
\begin{cases}
\pi_0 = (1-p) \pi_1 \\
\pi_1 = \pi_0 + (1-p) \pi_2 \\
\pi_i = p \pi_{i-1} + (1-p) \pi_{i+1}, \quad 1 < i < N-1 \\
\pi_{N-1} = p \pi_{N-2} + \pi_N \\
\pi_N = p \pi_{N-1}
\end{cases}
$$

We can obtain the following expressions for $\pi_1$ and $\pi_2$:



$$
\pi_1 = \frac{1}{1-p} \pi_0 = \frac{1}{p} \frac{p}{1-p} \pi_0 \\
\pi_2 = \frac{1}{1-p}\left(\frac{1}{1-p} - 1\right) \pi_0 = \frac{1}{p}\left(\frac{1}{1-p}\right)^2 \pi_0
$$

By induction, let's assume for some $k$: $\pi_k = \frac{1}{p}\left(\frac{1}{1-p}\right)^k \pi_0$. Then for $k+1$, we have:

$$
\pi_{k+1} = \frac{1}{1-p}\left(\frac{1}{p}\left(\frac{p}{1-p}\right)^k - \left(\frac{p}{1-p}\right)^{k-1}\right) \pi_0 = \frac{1}{p}\left(\frac{p}{1-p}\right)^{k+1} \pi_0
$$

This confirms that our assumption is valid for all $0 < i < N$.

Finally, we find that $\pi_N = p \pi_{N-1} = \left(\frac{p}{1-p}\right)^{N-1} \pi_0$.

Let's define $v_i$ as follows:

$$
v_i = \begin{cases}
1, & i = 0 \\
\frac{1}{p}\left(\frac{p}{1-p}\right)^i, & 0 < i < N \\
\left(\frac{p}{1-p}\right)^{N-1}, & i = N
\end{cases}
$$

So, the stationary distribution is given by: $\pi_i = \frac{v_i}{\sum v_i}$.

<u> THE GAMBLERâ€™S RUIN </u>

Consider the following chain that models the situation of a gambler in some casino:

image.png

#### In this chain, the transition probabilities are given by:
$$
\cases{p_{00} = 1\\p_{i;\,i-1}=q;\quad p_{i;\,i+1}=p\qquad \text{for }i=1;\,2;\,\dots }
$$
#### We assume that $p;\,q>0$.
#### Compute the run probability $h_i=P\left(\{\exists n\geq0;\,X_n=0\}:\,X_0=i\right)$. Show that even if the casino is fair, that is if $p=q$, the gambler probability to be ruined is 1.


Utilizing the provided transition probabilities: $p_{00}=1$, $p_{i,i-1}=q$, and $p_{i,i+1}=p$ for $i=1,2,\dots$, we are trying to compute the ruin probability:

$$
h_i = \mathbb{P}\big( \{ \exists n \ge 0, X_n = 0 \} | X_0 = i \big).
$$

We can express this probability as a sum of two terms:

$$
h_i = \mathbb{P}\big( \{ \exists n \ge 1, X_n = 0 \} | X_1 = i-1 \big)p_{i, i-1} + \mathbb{P}\big( \{ \exists n \ge 1, X_n = 0 \} | X_1 = i+1 \big)p_{i, i+1}
$$

Simplifying, we get:

$$
h_i = qh_{i-1}+ph_{i+1}
$$

This is a linear difference homogeneous equation. The boundary conditions are given by:    
- $h_0 = 1$ because if a gambler starts with no money, ruin is certain.
- $h_n = 0$ for $n \rightarrow \infty$ because a gambler with infinite money cannot be ruined.


To solve the linear homogeneous difference equation we propose a solution of the form $h_i = r^i$, and substitute this into our difference equation. This gives us

$$
r^i = qr^{i-1} + pr^{i+1}
$$

We can divide through by $r^{i-1}$ to get

$$
r = q + pr^2
$$

Rearranging terms gives the quadratic equation

$$
pr^2 - r + q = 0
$$

$pr^2-r+q=pr^2-(p+q)r+q=(rp-q)(r-1)=0\to r_1=1,r_2=\frac{q}{p}$

Now, we analyze two cases:

Case 1: $p \neq q$

In this scenario, the solution is of the form:

$$
h_i = A\big( \frac{q}{p} \big)^i + B1^i = A\big( \frac{q}{p} \big)^i + B
$$

Where $A$ and $B$ are constants. Using the boundary conditions, we solve for $A$ and $B$:

- For $h_0=A+B=1$ we get $B=1-A$.
- For $h_n=A\big( \frac{q}{p} \big)^n+1-A=0$ we find $A = \frac{1}{1-\big(\frac{q}{p}\big)^n}$ and subsequently $B = 1 - \frac{1}{1-\big(\frac{q}{p}\big)^n} = -\frac{\big(\frac{q}{p}\big)^n}{1-\big(\frac{q}{p}\big)^n}$.

Taking the limit as $n \to \infty$ gives $h_i=1$.

Case 2: $p = q$

In this case, the solution is of the form:

$$
h_i=(A+iB)1^i=A+iB
$$

Where $A$ and $B$ are constants. Using the boundary conditions, we solve for $A$ and $B$:

- For $h_0=A=1$ we get $A=1$.
- For $h_n = 1 + nB = 0$ we find $B = -\frac{1}{n}$.

$h_i=\lim_{n \to \infty} 1-\frac{i}{n}=1$

This demonstrates that, even under the assumption of a fair casino where $p = q$, the probability of the gambler being ruined is 1.

<u> HIDDEN MARKOV MODELS </u>

Welcome to the Emperor palace! The elder daughter of Probabil the Great, beautiful Princess Variance, reached the 2-pi-square age that is considered as a mean of a normal distribution for the age when maidens of the Empire get married. Today, the traditional matrimonial contest will be held in the Palace. The winner will marry the Princess.

The contest has ancient roots. When a princess is born, the Emperor assigns her a special servant. Every day from the day of the princess birth to the day of her 2-pi-square age her servant must visit four ponds in the Emperor Garden, in accordance to a Markov process, and catch one hundred fishes. The servant must record fish colors (red, blue, or green) and then return the fish back to the ponds. Each pond is strictly maintained and has its own proportion of fish of different colors. Every day, the Emperor mass media announced the

Ten days before the princess comes of age, her servant will put every caught fish into a transparent numbered jar and send it to the Palace. (It is assumed that taking out 1,000 fishes does not change the statistical properties of the ponds.) In order to win the Princess hand in marriage, a contestant has to guess as accurately as possible from which pond each fish came.

Our data is a sequence of observations $O$ of length $L = 1000$. Every data element is a color of a fish from a finite set of colors (or a finite alphabet of symbols). In our case the set contains three symbols: "red", "blue", and "green". Each fish was taken out of some pond, or we can say "a data point was emitted in some state q". A first-order Markov chain determines the
sequence of states, according to the following transition matrix:
$$
A=\left(\matrix{0.6& 0.05& 0.1 &0.25\\ 0.1& 0.05 &0.7& 0.15 \\ 0.05 &0.7& 0.1 &0.15 \\ 0.1& 0.1 &0.6 &0.2}\right)

Every state has its own discrete probability distribution for fish color. We shall call this distribution a symbol emission vector in $i$âˆ’th state. Collecting all vectors as columns of the matrix, we can get a symbol emission matrix:
$$
B=\left(\matrix{0.8& 0.2 &0.1& 0.5\\ 0.1& 0.7& 0.7& 0.3 \\ 0.1& 0.1 &0.2 &0.2}\right)\matrix{R\\G\\B}

```python
import numpy as np
from pandas import DataFrame as DF
```

## 1

```python
A = np.array([
    [0.6, 0.05, 0.1, 0.25],
    [0.1, 0.05, 0.7, 0.15],
    [0.05, 0.7, 0.1, 0.15],
    [0.1, 0.1, 0.6, 0.2]
])

B = np.array([
    [0.8, 0.2, 0.1, 0.5],
    [0.1, 0.7, 0.7, 0.3],
    [0.1, 0.1, 0.2, 0.2]
])

PI = np.array([0.3, 0.1, 0.4, 0.2])

data = []
```

```python
pond = np.random.choice([1, 2, 3, 4], p=PI)
for _ in np.arange(1000):
    p_colors = B[:, pond - 1]
    color = np.random.choice(["R", "G", "B"], p=p_colors)
    data.append({
        "pond": pond,
        "color": color
    })
    pond = np.random.choice([1, 2, 3, 4], p=A[pond - 1])

colors, counts = np.unique([e["color"] for e in data], return_counts=True)
print(f"Fish Count: {list(colors)} ~ {list(counts)}")

ponds, counts = np.unique([e["pond"] for e in data], return_counts=True)
print(f"Pond Count: {list(ponds)} ~ {list(counts)}")

COLORS = [e["color"] for e in data]
c = ["R", "G", "B"]
```

## 2

```python
def forward(PI, A, B, COLORS):
    alpha = np.zeros((len(COLORS), len(PI)), dtype=np.float64)
    for i in np.arange(alpha.shape[0]):
        color_id = c.index(COLORS[i])
        for j in np.arange(alpha.shape[1]):
            if i == 0:
                alpha[i, j] = PI[j] * B[color_id, j]
            else:
                alpha[i, j] = B[color_id, j] * np.sum(A.T[j] * alpha[i - 1])
    return alpha
```

## 3

```python
def backward(A, B, COLORS):
    beta = np.zeros((len(COLORS), len(ponds)), dtype=np.float64)

    for i in np.arange(beta.shape[0] - 1, -1, -1):
        for j in np.arange(beta.shape[1]):
            if i == 999:
                beta[i, j] = 1
            else:
                color_id = c.index(COLORS[i + 1])
                beta[i, j] = np.sum(B[color_id] * A[j] * beta[i + 1])
    return beta
```

## 4

```python
COLORS = [e["color"] for e in data]
color_id = c.index(COLORS[0])

alpha= forward(PI, A, B, COLORS)
p_forward = np.sum(alpha[-1])
print("Probability of the sequence of observations (forward):", p_forward)

beta = backward(A, B, COLORS)
p_backward = np.sum(B[color_id] * beta[0] * PI)
print("Probability of the sequence of observations (backward):", p_backward)
```

## 5

```python
def viterbi(PI, A, B, COLORS):
    delta = np.zeros((len(COLORS), len(ponds)), dtype=np.float64)
    psi = np.zeros((len(COLORS), len(ponds)), dtype=np.float64)

    for t in np.arange(delta.shape[0]):
        color_id = c.index(COLORS[t])
        for j in np.arange(delta.shape[1]):
            if t == 0:
                delta[t, j] = PI[j] * B[color_id, j]
                psi[t, j] = 0
            else:
                delta[t, j] = B[color_id, j] * np.max(delta[t - 1] * A.T[j])
                psi[t, j] = np.argmax(delta[t - 1] * A.T[j])

    P_star = np.max(delta[-1])
    q_star = np.zeros((len(COLORS)), dtype=np.float64)
    q_star[-1] = np.argmax(psi[-1])
    for i in np.arange(q_star.shape[0] - 2, -1, -1):
        q_star[i] = psi[i + 1, int(q_star[i + 1])]
        
    return q_star
```

## 6

```python
q_star = viterbi(PI, A, B, COLORS)
test = np.array([e["pond"] for e in data])
pred = q_star + 1

count = np.sum(test != pred)
accuracy = 1 - count / len(pred)
print("Accuracy of the most probable sequence of states:", accuracy)

comparison = DF({'Original': test, 'Predicted': pred})
print("Comparison of original and predicted states:")
comparison
```
